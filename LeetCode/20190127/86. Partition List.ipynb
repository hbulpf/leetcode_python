```python
class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution(object):
    def partition(self, head, x):
        """
        :type head: ListNode
        :type x: int
        :rtype: ListNode
        """
        if not head:
            return head
        h = ListNode(0)
        h.next = head
        p = head.next
        in1 = None
        in2 = None
        if head.val >= x:
            in1 = h
            in2 = head
        else:
            pre = head
            while p:
                if p.val >= x:
                    in1 = pre
                    in2 = p
                    break
                else:
                    p = p.next
                    pre = pre.next
        if not in1:
            return head
        p = in2.next
        pre = in2
        while p:
            if p.val < x:
                pre.next = p.next
                in1.next = p
                p.next = in2
                in1 = p
                p = pre.next if pre else None
            else:
                p = p.next
                pre = pre.next
        return h.next

head = ListNode(3)
head.next = ListNode(1)
head.next.next = ListNode(2)
b = Solution()
h = b.partition(head,3)
p = h

while p:
    print('#',p.val)
    p = p.next
```
