```python
class Solution:
    def solve(self, board):
        """
        :type board: List[List[str]]
        :rtype: void Do not return anything, modify board in-place instead.
        """
        if len(board) == 0:
            return
        r = len(board)
        c = len(board[0])
        a = [[1]*c for row in range(r)]
        if r <= 2 or c <= 2:
            return
        
        def find0(i, j):
            if a[i][j] == 0:
                return
            if board[i][j] == 'O':
                a[i][j] = 0
            else:
                return
            if i == 0 or i == r-1 or j == 0 or j == c-1:
                return
            find0(i,j-1)
            find0(i,j+1)
            find0(i+1,j)
            find0(i-1,j)
        
        if board[0][0] == 'O':
            a[0][0] = 0
        if board[0][c-1] == 'O':
            a[0][c-1] = 0
        if board[r-1][0] == 'O':
            a[r-1][0] = 0
        if board[r-1][c-1] == 'O':
            a[r-1][c-1] = 0
        for j in range(1, c-1):
            if board[0][j] == 'O':
                a[0][j] = 0
                find0(1,j)
            if board[r-1][j] == 'O':
                a[r-1][j] = 0
                find0(r-2,j)
        for i in range(1, r-1):
            if board[i][0] == 'O':
                a[i][0] = 0
                find0(i,1)
            if board[i][c-1] == 'O':
                a[i][c-1] = 0
                find0(i,c-2)
        for i in range(r):
            for j in range(c):
                if a[i][j] == 1:
                    board[i][j] = 'X'
        
             
        
```
