```python
class Solution(object):
    def ladderLength(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: int
        """
        if not endWord in wordList:
            return 0
        currentq = [beginWord]
        nextq = []
        n = len(wordList)
        m = 1
        
        def canchange(u, v):
            n = len(u)
            flag = 0
            for i in range(n):
                if u[i] != v[i]:
                    flag += 1
            if flag == 1:
                return True
            else:
                return False
        
        while m <= n:
            while currentq:
                t = currentq.pop(0)
                if canchange(t, endWord):
                    return m + 1
                else:
                    for v in wordList:
                        if canchange(t, v) and not v in nextq:
                            nextq.append(v)
            if not nextq:
                return 0
            else:
                currentq = nextq
                nextq = []
                m += 1
```

```python
class Solution(object):
    def ladderLength(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: int
        """
        import copy
        if not endWord in wordList:
            return 0
        if beginWord in wordList:
            wordList.pop(wordList.index(beginWord))
        currentq = [[beginWord, wordList]]
        nextq = []
        n = len(wordList)
        m = 1
        
        def canchange(u, v):
            n = len(u)
            flag = 0
            for i in range(n):
                if u[i] != v[i]:
                    flag += 1
            if flag == 1:
                return True
            else:
                return False
        
        while m <= n:
            while currentq:
                t = currentq.pop(0)
                if canchange(t[0], endWord):
                    return m + 1
                else:
                    for v in t[1]:
                        if canchange(t[0], v):
                            p = copy.copy(t[1])
                            p.pop(p.index(v))
                            nextq.append([v, p])
            if not nextq:
                return 0
            else:
                currentq = nextq
                nextq = []
                m += 1
```
