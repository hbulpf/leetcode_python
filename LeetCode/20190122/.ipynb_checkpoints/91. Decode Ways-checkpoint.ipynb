```python
class Solution:
    def numDecodings(self, s):
        """
        :type s: str
        :rtype: int
        """
        n = len(s)
        if n == 0 or s[0] == '0':
            return 0
        if n == 1:
            return 1
        a = 1
        if s[1] == '0':
            if s[0] in '12':
                b = 1
            else:
                b = 0
        else:
            if int(s[0:2]) in range(1,27):
                b = 2
            else:
                b = 1
        if n == 2:
            return b
        for i in range(1, n - 1):
            if s[i] != '0' and int(s[i:i+2]) in range(1,27):
                if s[i+1] == '0':
                    t = b
                    b = a
                    a = t
                else:
                    t = b
                    b = a + b
                    a = t
            else:
                if s[i+1] == '0':
                    return 0
                else:
                    a = b
        return b
```
